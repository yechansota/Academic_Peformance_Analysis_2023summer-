#### **Data Dictionary**
---
**School**: The name of the educational institution where the student is enrolled.

**Gender**: The biological gender of the student.

**Address**: Denotes whether the student's residence is located in an Urban (U) or Rural (R) area.

**Famsize**: Indicates the size of the student's family, categorized as either LE3 (below 3 members) or GT3 (over 3 members).

**Pstatus**: Reflects the living status of the parents in relation to the child. T signifies living together, while A indicates not living together.

**Medu**: The educational level of the student's mother, rated on a scale from 1 to 5.

**Fedu**: The educational level of the student's father, similarly rated on a scale from 1 to 5.

**Mjob**: The occupation of the student's mother.

**Fjob**: The occupation of the student's father.

**Reason**: The rationale behind the student's selection of the specific school.

**Guardian**: The individual responsible for the care and guardianship of the student.

**Traveltime**: The amount of time taken by the student to commute to the school.

**Studytime**: The number of hours dedicated to studying on a weekly basis.

**Failures**: The instances in which the student has experienced academic failure.

**Schoolsup**: Government-provided educational assistance for children.

**Famsup**: Support for the student's education from the family.

**Paid**: Participation in paid additional tutoring or educational services.

**Activities**: Involvement in extracurricular activities related to academic pursuits.

**Nursery**: Indicates whether the student utilizes nursery services.

**Higher**: The student's inclination towards pursuing higher education.

**Internet**: Availability of internet access in the student's household.

**Romantic**: The presence of a romantic relationship in the student's
life.

**Famrel**: The quality of family relationships, rated on a scale from 1 to 5.

**Freetime**: The amount of leisure time available to the student.

**Goout**: The frequency of the student's social interactions with peers.

**Dalc**: The student's daily alcohol consumption frequency.

**Walc**: The student's weekly alcohol consumption frequency.

**Health**: The student's self-reported health status.

**Absences**: The number of times the student has been absent from
classes.

**G1**: The academic grade achieved in the first year of study.

**G2**: The academic grade attained in the second year of study.

**G3**: The academic grade acquired in the third and final year of study.

### Data Cleansing for Modeliing
---

````python
# ▶ numeric, categorical value 나누기
numeric_list=[]
categoical_list=[]

for i in df.columns :
  if df[i].dtypes == 'O' :
    categoical_list.append(i)
  else :
    numeric_list.append(i)

print("categoical_list :", categoical_list)
print("numeric_list :", numeric_list)
````

````python
categoical_list.remove('Age_gp')
categoical_list.remove('flag')
````

> Note
---
* Due to the high correlation between G1 and G2, it is recommended to remove one of them from the analysis in order to minimize the potential for errors.

````python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

X=df.drop(['Age_gp', 'flag', 'G1', 'G2', 'G3'], axis=1)
Y=df['G3']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=123456)

print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
````

````python
from sklearn.preprocessing import LabelEncoder

for col in categoical_list:
    print(col)
    le = LabelEncoder()
    le.fit(list(x_train[col].values) + list(x_test[col].values))
    x_train[col] = le.transform(x_train[col])
    x_test[col] = le.transform(x_test[col])
````

### **Model Evaluation**
---


````python
rfr = RandomForestRegressor(n_estimators=500, max_depth=3,random_state=123456)
rfr.fit(x_train, y_train)
````

> Note
---
* It is necessary for the model evaluation to adjust Max and Estimate to reduce overfitting

````python
from sklearn.metrics import mean_absolute_error, r2_score
y_pred_train = rfr.predict(x_train)
y_pred_test = rfr.predict(x_test)


mae_train = mean_absolute_error(y_train, y_pred_train)
print('mae_train(mae): ', mae_train)
r2_train = r2_score(y_train, y_pred_train)
print('r2_score(r2): ', r2_train)
print('')
mae_test = mean_absolute_error(y_test, y_pred_test)
print('mae_test(mae): ', mae_test)
r2_test = r2_score(y_test, y_pred_test)
print('r2_score(r2): ', r2_test)
````

````python
# ▶ pd.options.display.float_format = '{:.5f}'.format
result = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred_test})
result['diff'] = result['Predicted Values'] - result['Real Values']
result.head(5)
````
